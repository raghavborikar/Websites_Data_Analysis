# Data Loading & Cleaning

```python
df['Website'] = df['Website'].str.replace('^www\.', '', regex=True)
df['Website'].value_counts()
```

```python
for col in ['Facebook_likes','Twitter_mentions','Google_pluses','Pinterest_pins','LinkedIn_mentions','StumbleUpon_views']:
    print(df[col].value_counts()) 
```

```python
df.replace('-', 0 , inplace=True)
```

```python
import re

def convert_to_numeric(value):
    if isinstance(value, str):
        if 'K' in value:
            return float(value.replace('K', '')) * 1000
        elif 'M' in value:
            return float(value.replace('M', '')) * 1000000
        elif 'B' in value:
            return float(value.replace('B', '')) * 1000000000
        else:
            try:
                return int(value)
            except ValueError:
                return value
    elif pd.isna(value):
        return int(value)
    else:
        return int(value)

for col in ['Facebook_likes','Twitter_mentions','Google_pluses','Pinterest_pins','LinkedIn_mentions','StumbleUpon_views']:
    df[col] = df[col].apply(convert_to_numeric)

for col in ['Facebook_likes','Twitter_mentions','Google_pluses','Pinterest_pins','LinkedIn_mentions','StumbleUpon_views']:
    print(df[col].value_counts())
```

```python
df['Avg_Daily_Pageviews'] = df['Avg_Daily_Pageviews'].astype(str).str.replace(' ', '')
df['Avg_Daily_Visitors'] = df['Avg_Daily_Visitors'].astype(str).str.replace(' ', '')
df[['Avg_Daily_Pageviews', 'Avg_Daily_Visitors']] = (df[['Avg_Daily_Pageviews', 'Avg_Daily_Visitors']].apply(pd.to_numeric, errors='coerce').fillna(0).astype('int64')                        )
```

```python
df
```

```python
for column in df.columns:
    if df[column].dtype in ['float64', 'int64']:
        website_means = df.groupby('Website')[column].transform('mean')
        
        df[column] = df[column].replace(0, pd.NA).fillna(website_means)

for column in df.columns:
    if df[column].dtype in ['float64', 'int64']:
        country_medians = df.groupby('country')[column].transform('median')
        
        df[column] = df[column].replace(0, pd.NA).fillna(country_medians)
```

```python
df['Country_Rank'].value_counts()
```

```python
df.drop(['Registrant','Reach_Day_percentage','Month_Average_Daily_Reach_percentage','Daily_Pageviews_percentage','Month_Average_Daily_Pageviews_percentage','Daily_Pageviews_per_user_percentage'],axis=1,inplace=True)
```

```python
df['Subnetworks'].value_counts()
```

```python
df['subnetwork_count'] = df['Subnetworks'].astype(str).apply(lambda x: len(x.split(',')))
```

```python
df['subnetwork_count'].fillna(1,inplace=True)
df.drop('Subnetworks',axis=1,inplace=True)
```

```python
df['Location'].isnull().sum()
```

```python
df['Location'].fillna('Unknown',inplace=True)
```

```python
df['Hosted_by'].fillna('Unknown',inplace=True)
```

```python
df['Registrar'].fillna('Unknown',inplace=True)
```

```python
df['Traffic_Rank'] = df['Traffic_Rank'].astype(str).str.replace(' ', '')
df['Traffic_Rank'] = df['Traffic_Rank'].apply(pd.to_numeric, errors='coerce')
df['Traffic_Rank'].fillna(df['Traffic_Rank'].median(), inplace=True)
```

```python
df[['Facebook_likes','Google_pluses','Twitter_mentions','Pinterest_pins','LinkedIn_mentions','StumbleUpon_views','Traffic_Rank']] = df[['Facebook_likes','Google_pluses','Twitter_mentions','Pinterest_pins','LinkedIn_mentions','StumbleUpon_views','Traffic_Rank']].astype(int)
```

```python
df.info()
```

```python
columns_to_update = ['Reach_Day', 'Month_Average_Daily_Reach', 'Daily_Pageviews','Month_Average_Daily_Pageviews', 'Daily_Pageviews_per_user']

for col in columns_to_update:
    df[col] = df.groupby('Website')[col].transform('mean').abs()
```

```python
df.drop('Status',axis=1,inplace=True)
```

```python
df['Status'] = 'Safe'

df.loc[(df['Location'] == 'Unknown') | (df['Hosted_by'] == 'Unknown') | (df['Registrar'] == 'Unknown'), 'Status'] = 'Risky'

df.loc[((df['Child_Safety']=='Unknown') & (df['Privacy']=='Unknown')), 'Status'] = 'Risky'

df['Status'].value_counts()
```

```python
df.to_csv('websites_data.csv', index=False)
```

```python
fig = go.Figure(data=[go.Sankey(
    node = dict(
      pad = 15,
      thickness = 20,
      line = dict(color = "blue", width = 0.5),
      label = df1['Privacy'].unique().tolist() + df1['Child_Safety'].unique().tolist(),
      color = "blue"
    ),
    link = dict(
      source = [df1['Privacy'].unique().tolist().index(val) for val in df1['Privacy']],
      target = [len(df1['Privacy'].unique()) + df1['Child_Safety'].unique().tolist().index(val) for val in df1['Child_Safety']],
      value = [1] * len(df1)
  ))])

fig.update_layout(title_text="Sankey Diagram between Privacy and Child Safety", font_size=10)
fig.show()
```

```python
import networkx as nx
import plotly.graph_objects as go

# Create the graph edges from your dataset
edges = list(zip(df1['Registrar'], df1['Location']))
G = nx.Graph()
G.add_edges_from(edges)

# Generate 3D node positions using spring layout
pos = nx.spring_layout(G, dim=3)  # Create 3D positions
for node, position in pos.items():
    G.nodes[node]['pos'] = position

# Extract node positions for plotting
x_nodes = [pos[i][0] for i in G.nodes()]  # x-coordinates of nodes
y_nodes = [pos[i][1] for i in G.nodes()]  # y-coordinates of nodes
z_nodes = [pos[i][2] for i in G.nodes()]  # z-coordinates of nodes

# Create edges
edge_x = []
edge_y = []
edge_z = []
for edge in G.edges():
    x0, y0, z0 = G.nodes[edge[0]]['pos']
    x1, y1, z1 = G.nodes[edge[1]]['pos']
    edge_x.extend([x0, x1, None])  # None creates breaks in lines
    edge_y.extend([y0, y1, None])
    edge_z.extend([z0, z1, None])

# Create Plotly traces
# Edge trace
edge_trace = go.Scatter3d(
    x=edge_x, y=edge_y, z=edge_z,
    mode='lines',
    line=dict(width=2, color='blue'),
    hoverinfo='none'
)

# Node trace
node_trace = go.Scatter3d(
    x=x_nodes, y=y_nodes, z=z_nodes,
    mode='markers',
    marker=dict(
        size=10,
        color='orange',
        opacity=0.8
    ),
    text=list(G.nodes()),  # Hover info for nodes
    hoverinfo='text'
)

# Create the 3D figure
fig = go.Figure(data=[edge_trace, node_trace],
                layout=go.Layout(
                    title='3D Network Diagram of Registrars and Locations',
                    showlegend=False,
                    scene=dict(
                        xaxis=dict(showbackground=False, title="X-axis"),
                        yaxis=dict(showbackground=False, title="Y-axis"),
                        zaxis=dict(showbackground=False, title="Z-axis")
                    )
                ))

# Show the figure
fig.show()
```
