```python
import re

def convert_to_numeric(value):
    if isinstance(value, str):
        if 'K' in value:
            return float(value.replace('K', '')) * 1000
        elif 'M' in value:
            return float(value.replace('M', '')) * 1000000
        elif 'B' in value:
            return float(value.replace('B', '')) * 1000000000
        else:
            try:
                return int(value)
            except ValueError:
                return value
    elif pd.isna(value):
        return int(value)
    else:
        return int(value)

for col in ['Facebook_likes','Twitter_mentions','Google_pluses','Pinterest_pins','LinkedIn_mentions','StumbleUpon_views']:
    df[col] = df[col].apply(convert_to_numeric)

for col in ['Facebook_likes','Twitter_mentions','Google_pluses','Pinterest_pins','LinkedIn_mentions','StumbleUpon_views']:
    print(df[col].value_counts())
```

```python
X = df3.drop(columns=['Website','Traffic_Rank','Location','Hosted_by','Registrar'])

categorical_columns = ['Status']
numerical_columns = ['Avg_Daily_Visitors', 'Avg_Daily_Pageviews', 'Trustworthiness', 'Child_Safety', 'Privacy', 'SM_PC1', 'SM_PC2', 'RS_PC1']

# Preprocess the features: One-hot encode categorical variables and scale numerical features
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_columns),  # Scale numerical columns
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_columns)  # One-hot encode categorical columns
    ]
)

# Create a pipeline to preprocess and apply DBSCAN
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),  # Preprocessing step
    ('dbscan', DBSCAN(eps=0.5, min_samples=10))  # DBSCAN clustering
])

# Fit the pipeline on the data
pipeline.fit(X)

# Get the cluster labels from DBSCAN
cluster_labels = pipeline.named_steps['dbscan'].labels_

# Add the cluster labels to the dataframe
df3['Cluster'] = cluster_labels

# Print out the results
print(df3[['Website', 'Cluster']].head())  # Show the website and cluster assignments

# Optionally, check how many clusters were found (DBSCAN labels clusters as 0, 1, ..., n, and -1 represents noise points)
num_clusters = len(set(cluster_labels)) - (1 if -1 in cluster_labels else 0)
print(f"Number of clusters (excluding noise): {num_clusters}")
```
